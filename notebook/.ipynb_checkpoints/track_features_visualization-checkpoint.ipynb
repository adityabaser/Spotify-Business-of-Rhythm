## Visualising track features over time

By analysing the popular songs over time and visualizing their features, we infer that the track features of most new songs converge or lie within a very specific range. Hence, if a new thriving artist wants his music to be popular in this age, he/she should stay within this range of the features. 

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as spst
from scipy.optimize import minimize
import scipy.special as spsp

import warnings
warnings.simplefilter("ignore")
```

```python
session = pd.read_csv('../data/session_info.csv')
Track_feature = pd.read_csv('../data/track_features.csv')
```

```python
data = session.merge(Track_feature, left_on='track_id_clean', right_on='track_id')
data.head()
```

```python
filter = data.groupby(["release_year"]).mean().reset_index()
a = filter.iloc[:,-25:]
b = filter.iloc[:,0:1]
c = filter.iloc[:,16:17]
songs = pd.concat([b,c,a],axis=1)
```

```python
for j in songs.columns[1:]:
  for i in range(0,len(songs)):
    plt.scatter(songs["release_year"],songs[j],c="red")
    plt.grid()
    plt.title(j+" with time")
  plt.show()
```
